```python
#sequential search
def seq_search(a,b,n):
    """
    a를 정수의 상한으로 가지는 b개의 집합중
    n이 존재하는지, 존재한다면 몇번 째 위치에 있는지
    """
    count = 0
    judge = True #추후 검색값을 갖는 원소 판단
    seq_list = np.random.randint(a+1,size = b) #a+1이 상한선, size = b
    print(f'만들어진 리스트 {seq_list}')
    for i in range(0,len(seq_list)):
        if seq_list[i] == n:
            print(f'검색값 {n}은 seq_list[{i}]에 있습니다.')
            judge = False
            count +=1
    if judge == True:
        print(f'검색값 {n}은 {seq_list}에 없습니다' )
    return print(count)
```
